```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# 1. Connect to your Docker Database
# We use the connection string we tested earlier
db_url = "postgresql://admin:12345@127.0.0.1:5432/medical_safety_db"
engine = create_engine(db_url)

# 2. Query the Data
# We ask: "Get me Aztreonam, its targets, and the proteins those targets talk to."
sql = """
SELECT 
    d.generic_name AS source, 
    t.target_uniprot_id AS target,
    'Drug-Target' as type
FROM drugs d
JOIN drug_targets t ON d.id = t.drug_id
WHERE d.generic_name = 'aztreonam'

UNION

SELECT 
    ppi.protein_a_uniprot AS source,
    ppi.protein_b_uniprot AS target,
    'Protein-Protein' as type
FROM protein_interactions ppi
WHERE ppi.protein_a_uniprot IN (
    SELECT t.target_uniprot_id 
    FROM drugs d 
    JOIN drug_targets t ON d.id = t.drug_id 
    WHERE d.generic_name = 'aztreonam'
)
"""

# Load data into Python
df = pd.read_sql(sql, engine)
print(f"Loaded {len(df)} connections for visualization.")

# 3. Draw the Graph
# Create a network object
G = nx.from_pandas_edgelist(df, source='source', target='target', edge_attr='type')

plt.figure(figsize=(12, 8)) # Make the image big

# Layout (Spring layout spreads nodes out naturally)
pos = nx.spring_layout(G, seed=42)

# Draw the Drug (Red)
drugs = [n for n in G.nodes if n == 'aztreonam']
nx.draw_networkx_nodes(G, pos, nodelist=drugs, node_color='#ff4b4b', node_size=2000, label='Drug')

# Draw the Proteins (Blue)
proteins = [n for n in G.nodes if n != 'aztreonam']
nx.draw_networkx_nodes(G, pos, nodelist=proteins, node_color='#4b7bff', node_size=800, label='Protein')

# Draw the Lines
nx.draw_networkx_edges(G, pos, width=1, alpha=0.5)

# Add Labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif", font_color="black")

plt.title("Biological Network: How 'Aztreonam' interacts with the body", fontsize=15)
plt.legend()
plt.axis('off') # Hide the box border
plt.show()
```
